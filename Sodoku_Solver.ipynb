```python
import numpy as np
```

```python
board = [
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9]  
]
board = np.array(board)
```

```python
def print_board():
    
    square_len = len(board[0]) // 3 * 5
    tab_len = (len(board) - 1) // 3 * 3
    
    print("{}{}".format(
            '-' * square_len, 
            '-' * tab_len
        )
    )
    for i in range(len(board)):
        if i > 0 and i % 3 == 0: print()
        for j in range(len(board[i])):
            if j > 0 and j % 3 == 0: print("\t", end="")
            print("{} ".format(board[i,j]), end="")
        print()
    print("{}{}".format(
            '-' * square_len, 
            '-' * tab_len
        )
    )
```

```python
print_board()
```

```python
def possible(x: int, y: int, n: int)->bool:
    global board
    
    sq_x = 3 * (x // 3)
    sq_y = 3 * (y // 3)
    
    # if in square return false
    if n in board[sq_y:sq_y + 3, sq_x:sq_x + 3]: return False
    
    # if in row return false
    if n in board[y]: return False
    
    # if in column return false
    if n in np.transpose(board)[x]: return False
    
    return True
    
```

```python
def solve():
    global board
    
    for x in range(len(board[0])):
        for y in range(len(board)):
            if board[y,x] == 0:
                for n in range(1, 10):
                    if possible(x, y, n):
                        board[y,x] = n
                        solve()
                        board[y,x] = 0
                return
    print(board)
                
            
```

```python
solve()
```
